```python
from pathlib import Path

data_dir = Path('/Users/misha/data')  # Change this to the directory where you store the data

demo_images_dir = data_dir / 'demo_images'
all_bboxes = demo_images_dir / 'all_bboxes'
suspect_bboxes = demo_images_dir / 'suspect_bboxes'
relabeled_bboxes = demo_images_dir / 'relabeled_bboxes'

all_images_names = [f.name for f in all_bboxes.iterdir()]
```

```python
len(all_images_names)
```

```python
import matplotlib.pyplot as plt

for image_name in all_images_names[:4]:
    all_image = plt.imread(all_bboxes / image_name)
    suspect_image = plt.imread(suspect_bboxes / image_name)
    relabeled_image = plt.imread(relabeled_bboxes / image_name)

    fig, ax = plt.subplots(1, 3, figsize=(15, 5))
    ax[0].imshow(all_image)
    ax[0].set_title('All bboxes')
    ax[1].imshow(suspect_image)
    ax[1].set_title('Suspect bboxes')
    ax[2].imshow(relabeled_image)
    ax[2].set_title('Relabeled bboxes')

    plt.show()
```
